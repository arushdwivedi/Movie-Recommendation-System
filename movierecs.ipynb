```python
import numpy as np 
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

#pre processing:


```python
movies_dataframe=pd.read_csv('dataset.csv')
```

```python
movies_dataframe.head()
```

```python
selected_features=['genres','keywords','tagline','cast','director']
for feature in selected_features:
    movies_dataframe[feature]= movies_dataframe[feature].fillna('')
    
```

COMBINING THE 5 RELEVANT FEATURES:

```python
combined_features= movies_dataframe['genres']+movies_dataframe['keywords']+movies_dataframe['tagline']+movies_dataframe['cast']+movies_dataframe['director']
combined_features.head()
```

converting text data to a numerical equivalent:

```python
vectorizer=TfidfVectorizer()
featurevectors=vectorizer.fit_transform(combined_features)
print(featurevectors)
```

COSINE SIMILARITY:

```python
similarity=cosine_similarity(featurevectors)
print(similarity)
```

```python
print(similarity.shape)
```

getting movie to base recommendations on from the user:

```python
movie_name=input("Enter your favourite movie's name: ")
```

```python
list_of_titles=movies_dataframe['title'].tolist()
```

```python
find_close_match=difflib.get_close_matches(movie_name,list_of_titles)
print(find_close_match)
```

```python
closestmatch=find_close_match[0] # closest bestest match in list
```

```python
index_of_movie=movies_dataframe[movies_dataframe.title==closestmatch]['index'].values[0]
print(index_of_movie)
```

```python
similarityscore=list(enumerate(similarity[index_of_movie]))
print(similarityscore)
```

```python
sorted_similar_movies=sorted(similarityscore,key= lambda x:x[1],reverse= True)
print(sorted_similar_movies)
```

print name of similar movies:

```python
print("Movies we suggest you to watch:")
i=1
for movie in sorted_similar_movies:
    index= movie[0]
    titleofmovie=movies_dataframe[movies_dataframe.index==index]['title'].values[0]
    if(i<15):
        print(i,'.',titleofmovie)
        i=i+1
```
